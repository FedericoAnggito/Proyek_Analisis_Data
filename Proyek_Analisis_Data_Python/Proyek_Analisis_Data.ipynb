```python
all_df = pd.merge(
    left=day_df,
    right=hour_df,
    how="left",
    left_on="dteday",
    right_on="dteday"
)
all_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Visualisasi cuaca terhadap jumlah peminjaman sepeda
plt.figure(figsize=(10, 6))
sns.barplot(x="weathersit_x", y="cnt_x", data=all_df, ci=None)
plt.title("Pengaruh Cuaca Terhadap Jumlah Peminjaman Sepeda")
plt.xlabel("Cuaca")
plt.ylabel("Jumlah Peminjaman")
plt.show()
```

```python
# Visualisasi cuaca terhadap jumlah peminjaman sepeda
plt.figure(figsize=(10, 6))
sns.barplot(x="weathersit_y", y="cnt_y", data=all_df, ci=None)
plt.title("Pengaruh Cuaca Terhadap Jumlah Peminjaman Sepeda")
plt.xlabel("Cuaca")
plt.ylabel("Jumlah Peminjaman")
plt.show()
```

### Pertanyaan 2:

```python
# Visualisasi hubungan antara tingkat kelembaban udara dan jumlah peminjaman sepeda
plt.figure(figsize=(10, 6))
sns.scatterplot(x="hum_x", y="cnt_x", data=all_df)
plt.title("Hubungan Antara Kelembaban Udara dan Jumlah Peminjaman Sepeda")
plt.xlabel("Kelembaban Udara")
plt.ylabel("Jumlah Peminjaman")
plt.show()
```

```python
# Visualisasi hubungan antara tingkat kelembaban udara dan jumlah peminjaman sepeda
plt.figure(figsize=(10, 6))
sns.scatterplot(x="hum_y", y="cnt_y", data=all_df)
plt.title("Hubungan Antara Kelembaban Udara dan Jumlah Peminjaman Sepeda")
plt.xlabel("Kelembaban Udara")
plt.ylabel("Jumlah Peminjaman")
plt.show()
```

  Berdasarkan visualisasi data yang ada di bar plot menunjukkan bahwa jumlah peminjaman sepeda dapat bervariasi tergantung kondisi cuaca.


- conclution pertanyaan 2 : Apakah ada hubungan antara tingkat kelembaban udara dan jumlah peminjaman sepeda?

  Berdasarkan scatter plot tidak menunjukkan pola yang jelas. Tidak ada korelasi yang bisa diidentifikasi. Ini menunjukkan bahwa tingkat kelembaban udara tidak mempengaruhi jumlah peminjaman sepeda.
